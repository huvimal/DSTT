```python
from PIL import Image

img = Image.open("traidat.jpg")  # Tên file đúng đã upload

# Kiểm tra thông tin ảnh
print("Kích thước gốc:", img.width, "x", img.height)
print("Chế độ màu:", img.mode)

# Chuyển sang RGB nếu cần
if img.mode != 'RGB':
    img = img.convert("RGB")

# Resize ảnh giảm 50%
new_width = int(img.width / 2)
new_height = int(img.height / 2)

try:
    new_img = img.resize((new_width, new_height), Image.LANCZOS)
except AttributeError:

    from PIL import Image
    new_img = img.resize((new_width, new_height), Image.Resampling.LANCZOS)

# Lưu lại
new_img.save("traidat_small.png")
print("Đã lưu ảnh sau khi giảm kích thước thành công!")
from PIL import Image, ImageDraw


input_image = Image.open('traidat.jpg')
input_pixels = input_image.load()
# Định nghĩa kernel làm mờ dạng hộp 3x3.
# Mỗi phần tử trong kernel là 1/9, phản ánh việc lấy trung bình của 9 pixel.
box_kernel = [[1/9, 1/9, 1/9],
              [1/9, 1/9, 1/9],
              [1/9, 1/9, 1/9]]

# Gán kernel đã định nghĩa.
kernel = box_kernel

# Tính toán offset để xử lý biên.
# Offset là bán kính của kernel, ví dụ với kernel 3x3, offset là (3-1)/2 = 1.
offset = len(kernel) // 2
# Tạo một ảnh mới với cùng chế độ màu và kích thước với ảnh đầu vào.
output_image = Image.new("RGB", input_image.size)

# Tạo đối tượng ImageDraw để vẽ các pixel lên ảnh đầu ra.
draw = ImageDraw.Draw(output_image)
# Duyệt qua từng pixel của ảnh đầu vào (trừ các pixel biên).
# x và y là tọa độ của pixel hiện tại.
for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        # Khởi tạo accumulator cho các kênh màu RGB.
        acc = [0, 0, 0]

        # Duyệt qua các phần tử của kernel.
        for a in range(len(kernel)):
            for b in range(len(kernel)):
                # Tính toán tọa độ của pixel lân cận.
                # xn, yn là tọa độ của pixel xung quanh trong phạm vi kernel.
                xn = x + a - offset
                yn = y + b - offset

                # Lấy giá trị pixel (R, G, B) của pixel lân cận.
                pixel = input_pixels[xn, yn]

                # Tích lũy giá trị màu của từng kênh (R, G, B)
                # bằng cách nhân với giá trị tương ứng trong kernel.
                acc[0] += pixel[0] * kernel[a][b]  # Kênh đỏ
                acc[1] += pixel[1] * kernel[a][b]  # Kênh xanh lá
                acc[2] += pixel[2] * kernel[a][b]  # Kênh xanh dương

        # Vẽ điểm ảnh (x, y) trên ảnh đầu ra với giá trị màu đã làm mờ.
        # Chuyển đổi giá trị tích lũy sang kiểu số nguyên.
        draw.point((x, y), (int(acc[0]), int(acc[1]), int(acc[2])))
# Lưu ảnh đã làm mờ vào một file mới.
```

```python
#Bài tập chương 7
#Câu 1
import numpy as np
from PIL import Image

def gaussian_blur(image_path, output_path):
    """
    Thực hiện làm mờ Gauss trên một ảnh.

    Args:
        image_path (str): Đường dẫn đến ảnh đầu vào.
        output_path (str): Đường dẫn để lưu ảnh đã làm mờ.
    """
    try:
        img = Image.open(image_path).convert('L')  # Mở ảnh và chuyển sang ảnh xám
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp ảnh tại {image_path}")
        return
    except Exception as e:
        print(f"Lỗi khi mở ảnh: {e}")
        return

    img_np = np.array(img, dtype=np.float32)

    # Định nghĩa ma trận nhân Gaussian
    gaussian_kernel = np.array([
        [1, 4, 6, 4, 1],
        [4, 16, 24, 16, 4],
        [6, 24, 36, 24, 6],
        [4, 16, 24, 16, 4],
        [1, 4, 6, 4, 1]
    ], dtype=np.float32) / 256.0

    # Lấy kích thước ảnh và kernel
    img_height, img_width = img_np.shape
    kernel_height, kernel_width = gaussian_kernel.shape

    # Tính toán padding cần thiết để đầu ra có cùng kích thước với đầu vào
    pad_h = kernel_height // 2
    pad_w = kernel_width // 2

    # Thêm padding vào ảnh
    # Sử dụng 'edge' để lặp lại giá trị biên, giúp xử lý các pixel gần biên ảnh tốt hơn
    padded_img = np.pad(img_np, ((pad_h, pad_h), (pad_w, pad_w)), mode='edge')

    # Khởi tạo ma trận để lưu ảnh đầu ra
    output_img_np = np.zeros_like(img_np, dtype=np.float32)

    # Thực hiện tích chập
    for i in range(img_height):
        for j in range(img_width):
            # Lấy vùng lân cận của pixel hiện tại
            # Kích thước vùng lân cận bằng kích thước kernel
            neighborhood = padded_img[i:i + kernel_height, j:j + kernel_width]

            # Tính tổng tích (dot product) của vùng lân cận và kernel
            output_img_np[i, j] = np.sum(neighborhood * gaussian_kernel)

    # Đảm bảo giá trị pixel nằm trong khoảng [0, 255]
    output_img_np = np.clip(output_img_np, 0, 255)

    # Chuyển đổi mảng numpy trở lại đối tượng ảnh PIL và lưu
    output_img = Image.fromarray(output_img_np.astype(np.uint8))
    output_img.save(output_path)
    print(f"Ảnh đã làm mờ Gauss được lưu tại: {output_path}")

gaussian_blur("input_image.png", "output_gaussian.png")
```

```python
#Câu 2
import numpy as np
from PIL import Image

def apply_edge_detection(image_path, output_path_a, output_path_b):
    """
    Thực hiện phát hiện cạnh bằng các bộ lọc Laplacian đã cho.

    Args:
        image_path (str): Đường dẫn đến ảnh đầu vào.
        output_path_a (str): Đường dẫn để lưu ảnh sau khi áp dụng kernel a.
        output_path_b (str): Đường dẫn để lưu ảnh sau khi áp dụng kernel b.
    """
    try:
        img = Image.open(image_path).convert('L')  # Mở ảnh và chuyển sang ảnh xám
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp ảnh tại {image_path}")
        return
    except Exception as e:
        print(f"Lỗi khi mở ảnh: {e}")
        return

    img_np = np.array(img, dtype=np.float32)

    # Định nghĩa các kernel Laplacian
    kernel_a = np.array([
        [0, -1, 0],
        [-1, 4, -1],
        [0, -1, 0]
    ], dtype=np.float32)

    kernel_b = np.array([
        [-1, -1, -1],
        [-1, 8, -1],
        [-1, -1, -1]
    ], dtype=np.float32)

    # Danh sách các kernel để duyệt qua
    kernels = [("Kernel_A", kernel_a, output_path_a),
               ("Kernel_B", kernel_b, output_path_b)]

    img_height, img_width = img_np.shape

    # Kích thước kernel luôn là 3x3 cho Laplacian
    kernel_height, kernel_width = 3, 3

    pad_h = kernel_height // 2
    pad_w = kernel_width // 2

    padded_img = np.pad(img_np, ((pad_h, pad_h), (pad_w, pad_w)), mode='edge')

    for name, current_kernel, output_path in kernels:
        print(f"Đang xử lý với {name}...")
        output_img_np = np.zeros_like(img_np, dtype=np.float32)

        for i in range(img_height):
            for j in range(img_width):
                neighborhood = padded_img[i:i + kernel_height, j:j + kernel_width]
                output_img_np[i, j] = np.sum(neighborhood * current_kernel)

        # Bộ lọc Laplacian có thể tạo ra giá trị âm hoặc lớn hơn 255.
        # Để hiển thị, chúng ta thường chuẩn hóa lại về khoảng [0, 255].
        # Một cách phổ biến là chuẩn hóa tuyến tính.
        output_img_np_normalized = np.interp(output_img_np,
                                             (output_img_np.min(), output_img_np.max()),
                                             (0, 255))

        # Hoặc bạn có thể chỉ đơn giản là cắt và sau đó chuyển sang uint8 nếu muốn
        # output_img_np_normalized = np.clip(output_img_np, 0, 255)

        output_img = Image.fromarray(output_img_np_normalized.astype(np.uint8))
        output_img.save(output_path)
        print(f"Ảnh phát hiện cạnh với {name} được lưu tại: {output_path}")

# Ví dụ sử dụng:
# Sử dụng ảnh đã tạo từ Câu 1 hoặc một ảnh khác.
apply_edge_detection("input_image.png", "output_edge_a.png", "output_edge_b.png")
```
